# **Task 1 : Conceptual Understanding**

#### 1 What is the difference between "Love" and "love" in NLP?
In case of Love(upper case L) and love(lower case l), the computer considers them as two separate and different tokens if the case is not normalized.
It causes duplication in vocabulary and confusion for the models. This increases the vocabulary's size and reduces the model's efficiency. Without lowercasing the model wont know the difference between Love and love

#### 2 What happens if stopwords are not removed?
Stopwords are common words like The, is, at, and (It can be considered as noisy data). In case the stopwords are not removed, the vocabulary size will grow unnecessarily. The models will waste time and effort learning the meaning of these patterns which dont mean anything. It reduces accuracy in tasks like clustering and classification. Importatan words will get overshadowed by this noisy data.

#### 3 Provide two real-world scenarios where removing stopwords can be harmful.
1 Sentiment analysis: In case of sentiment analysis, 
  - Words like not or no completely change meaning.
  - Example: “I am not happy” -> if not is removed, it becomes “I am happy”, which flips the sentiment.
  
2 Legal/medical documents:
  - Stopwords may carry structural meaning.
  - Example: 
      - “without liability” -> removing without changes the meaning to “liability”, which is dangerous in legal contexts.
      - "Does not have cancer" -> if not is removed, it becomes "Does have cancer", which gives false positives or negatives.

#### 4 Explain the difference between stemming and lemmatization.
1 Stemming:
  - It is a rule-based process that chops off word endings. It is fast but can sometimes be inaccurate.
  - Example: “running” becomes “run” and “better” becomes “bet”.
  
2 Lemmatization:
  - This uses dictionaries and grammar rules to find the proper base form of the word. This is more accurate but slower.
  - Example: “running” becomes “run”, “better” becomes “good”.

# **Task 2: Build Advanced Preprocessing Function**

```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    if text == "" or not isinstance(text, str):
        return [], ""

    # Convert to lowercase
    text = text.lower()
    
    # Remove Urls and Emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Remove repeated letters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    tokens = text.split()
    
    # Remove very small tokens except "no" and "not"
    tokens = [t for t in tokens if len(t) > 2 or t in ["no", "not"]]
    
    clean_sentence = " ".join(tokens)
    
    return tokens, " ".join(tokens)
```

## **Task 3: Stress Testing**

```python
samples = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []
for s in samples:
    tokens, cleaned = preprocess_text(s)
    print(f"Original: {s}")
    print(f"Tokens: {tokens}")
    print(f"Cleaned: {cleaned}\n")
    results.append(tokens)
```

## **Task 4: Token Analytics**

```python
def token_stats(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens)/total if total > 0 else 0
    return total, unique, avg_len

results = []
for s in samples:
    tokens, cleaned = preprocess_text(s)
    total, unique, avg_len = token_stats(tokens)
    results.append({
        "original": s,
        "cleaned": cleaned,
        "tokens": tokens,
        "total_tokens": total,
        "unique_tokens": unique,
        "avg_token_length": round(avg_len, 2)
    })

for r in results:
    print(f"Original: {r['original']}")
    print(f"Cleaned: {r['cleaned']}")
    print(f"Tokens: {r['tokens']}")
    print(f"Total: {r['total_tokens']}, Unique: {r['unique_tokens']}, AvgLen: {r['avg_token_length']}\n")

# Analysis answers
most_noise = samples[-2]   
most_meaningful = samples[-1] 
print("Sentence with most noise:", most_noise)
print("Sentence with most meaningful tokens:", most_meaningful)
```

## **Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = [] #combine all tokens
for s in samples:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)
freq = Counter(all_tokens) 

print("Top 10:", freq.most_common(10))
print("Least 5:", freq.most_common()[-5:])
```

## **Task 6: Build Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

pipeline_output = full_pipeline(samples)
pipeline_output
```

## **Task 7: Error Handling**

```python
def preprocess_text(text):
    import re

    # This handles empty string
    if not text or text.strip() == "":
        return [], ""

    # Lowercase
    text = text.lower()
    
    # Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters (reduce to 2 max)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    # Remove punctuation and emojis (non-word characters)
    text = re.sub(r'[^\w\s]', ' ', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Tokenize
    tokens = text.split()
    
    # Remove very short tokens except "no" and "not"
    tokens = [t for t in tokens if len(t) > 2 or t in ["no", "not"]]
    
    return tokens, " ".join(tokens)

# --- Error Handling Tests ---
test_cases = [
    "",                  # Empty string
    "😂😂😂😂",        # Only emojis
    "1234567890",        # Only numbers
    "Hello, it's a nice day today!!!",   # Normal case
]

for case in test_cases:
    tokens, cleaned = preprocess_text(case)
    print(f"Original: {case!r}")
    print(f"Tokens: {tokens}")
    print(f"Cleaned: {cleaned}\n")
```
